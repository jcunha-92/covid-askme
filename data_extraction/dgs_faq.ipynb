# Extração da FAQ da DGS

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
import json
```

```python
url = 'https://covid19.min-saude.pt/perguntas-frequentes'
request = requests.get(url)
soup = BeautifulSoup(request.text, 'html.parser')
```

```python
#Get questions and answers

data = [] 
for elem in soup.find_all(True):
     temp = {}
     if elem.get('class') == ['has-text-color']:
          category = elem.get_text()
     if elem.get('class') == ['wp-block-spms-accordion-item', 'c-accordion__item', 'js-accordion-item', 'no-js']:
          temp['question'] = re.findall('>(.*)<', str(elem.find_all('h4')))[0]
          temp['answer'] = str(elem.find_all('p')).replace('[<p>','').replace('</p>]','').replace('<p>','').replace('</p>, ','\n')
          temp['category'] = category
     # discard empty data     
     if temp:     
          data.append(temp)
```

```python
#Dump json file  

import json
with open('../data/FAQ_DGS.json', 'w') as f:
    json.dump(data, f)     
```
