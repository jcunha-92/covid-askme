# Extração da FAQ Estamos on

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
import json
```

```python
url = 'https://covid19estamoson.gov.pt/faq/'
request = requests.get(url)
soup = BeautifulSoup(request.text, 'html.parser')
```

```python
def merge_two_dicts(x, y):
    z = x.copy()  
    z.update(y)  
    return z
          
questions = []
for div in soup.find_all('div'):
     temp = {}
     if div.get('class') in [['tab-pane', 'fade', 'show', 'active'], ['tab-pane', 'fade']]:
           category = div.get('id')
     if div.get('class') in [['card-header']]:      
           temp['question'] = div.get_text()
           temp['category'] = category
     # discard empty records
     if temp:
          questions.append(temp)


answers = []
for div in soup.find_all('div'):
     temp = {}
     if div.get('class') in [['card-body']]:
          temp['answer'] = div.get_text()
     # discard empty records     
     if temp:
          answers.append(temp)
          
# zip questions and answers
zip_data = list(zip(questions, answers))
# merge data
data = [merge_two_dicts(question, answer) for question, answer in zip_data]
```
